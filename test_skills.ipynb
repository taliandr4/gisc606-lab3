```python
import os

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import rioxarray as rxr

import earthpy as et

sns.set(font_scale=1.5, style="white")
```

```python
et.data.get_data("colorado-flood")
os.chdir(os.path.join(et.io.HOME, 'earth-analytics', 
                      'data'))
```

```python
lidar_dem_path = os.path.join("colorado-flood",
                              "spatial",
                              "boulder-leehill-rd",
                              "pre-flood",
                              "lidar",
                              "pre_DTM_hill.tif")
```

```python
if not os.path.exists(lidar_dem_path):
    raise FileNotFoundError(f"File '{lidar_dem_path}' not found. Please check the path.")
```

```python
lidar_dem = rxr.open_rasterio(lidar_dem_path)
```

```python
lidar_dem.shape
```

```python
f, ax = plt.subplots(figsize=(10, 5))
lidar_dem.plot(cmap="Greys_r",
                 ax=ax)
ax.set_title("Lidar Digital Terrain Model Overlayed On Top of a Hillshade \n Boulder Flood 2013")
ax.set_axis_off()
plt.show()
```
